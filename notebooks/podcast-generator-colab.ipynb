## استيراد المكتبات

```
import os
import numpy as np
from gtts import gTTS
from pydub import AudioSegment
from pydub.utils import mediainfo
import librosa
import soundfile as sf

print("✅ جميع المكتبات تم تحميلها بنجاح!")
```

## الخطوة 2️⃣: دالة تحويل النص إلى صوت

```
def text_to_speech_ar(text, filename, language='ar', slow=False):
    """
    تحويل النص إلى صوت بالعربية باستخدام Google Text-to-Speech
    
    المعاملات:
    - text: النص المراد تحويله
    - filename: اسم الملف الصوتي المراد حفظه
    - language: اللغة (ar للعربية، en للإنجليزية)
    - slow: تبطيء السرعة
    """
    try:
        tts = gTTS(text=text, lang=language, slow=slow)
        tts.save(filename)
        print(f"�� تم حفظ الملف: {filename}")
        return filename
    except Exception as e:
        print(f"❌ حدث خطأ: {e}")
        return None

print("✅ تم تعريف دالة تحويل النص إلى صوت")
```

## جرب الدالة الآن! 🎤

```
# مثال: تحويل نص عربي إلى صوت
sample_text = "مرحبا! هذا بودكاست من جوجل كولاب"
output_file = "podcast_sample.mp3"

text_to_speech_ar(sample_text, output_file)

# عرض الملف الصوتي
from IPython.display import Audio, display
display(Audio(output_file))
```

## الخطوة 3️⃣: دمج ملفات صوتية متعددة

```
def merge_audio_files(audio_files, output_name="podcast_merged.mp3"):
    """
    دمج ملفات صوتية متعددة في ملف واحد
    
    المعاملات:
    - audio_files: قائمة بأسماء الملفات الصوتية
    - output_name: اسم الملف النهائي
    """
    combined = AudioSegment.empty()
    
    for audio_file in audio_files:
        if os.path.exists(audio_file):
            sound = AudioSegment.from_mp3(audio_file)
            combined += sound
            print(f"✅ تم إضافة: {audio_file}")
        else:
            print(f"❌ الملف غير موجود: {audio_file}")
    
    # حفظ الملف المدمج
    combined.export(output_name, format="mp3")
    print(f"✅ تم حفظ الملف المدمج: {output_name}")
    return output_name

print("✅ تم تعريف دالة دمج الملفات الصوتية")
```

## الخطوة 4️⃣: رفع ملفاتك من حاسوبك

## الخطوة 5️⃣: تنزيل الملفات

```
def download_file(filename):
    """
    تنزيل ملف من Colab إلى حاسوبك
    """
    if os.path.exists(filename):
        files.download(filename)
        print(f"✅ بدأ تنزيل: {filename}")
    else:
        print(f"❌ الملف غير موجود: {filename}")

# مثال:
# download_file("podcast_sample.mp3")
print("📥 دالة التنزيل جاهزة للاستخدام")
```

## سير العمل الكامل 🔄

```
# 1️⃣ إنشاء نصوص متعددة
podcast_scripts = [
    "مرحبا! أنا المقدم الأول",
    "ومرحبا! أنا المقدم الثاني",
    "اليوم سنتحدث عن موضوع مهم جدا"
]

# 2️⃣ تحويل كل نص إلى صوت
audio_files = []
for i, script in enumerate(podcast_scripts, 1):
    filename = f"part_{i}.mp3"
    text_to_speech_ar(script, filename)
    audio_files.append(filename)

# 3️⃣ دمج الملفات
final_podcast = merge_audio_files(audio_files, "podcast_final.mp3")

# 4️⃣ عرض الملف النهائي
from IPython.display import Audio, display
print("\n🎧 استمع إلى البودكاست النهائي:")
display(Audio(final_podcast))
```
