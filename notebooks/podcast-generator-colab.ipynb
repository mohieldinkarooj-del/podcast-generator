## الخطوة 2️⃣: دالة تحويل النص إلى صوت

```
from gtts import gTTS
import os

def text_to_speech_ar(text, filename, language='ar', slow=False):
    """
    تحويل النص إلى صوت بالعربية باستخدام Google Text-to-Speech
    """
    try:
        tts = gTTS(text=text, lang=language, slow=slow)
        tts.save(filename)
        print(f"✅ تم حفظ الملف بنجاح: {filename}")
        return filename
    except Exception as e:
        print(f"❌ حدث خطأ أثناء تحويل النص: {e}")
        return None

print("✅ دالة text_to_speech_ar جاهزة للعمل")
```

## الخطوة 3️⃣: دمج ملفات صوتية متعددة

```
def merge_audio_files(audio_files, output_name="podcast_merged.mp3"):
    """
    دمج ملفات صوتية متعددة في ملف واحد

    المعاملات:
    - audio_files: قائمة بأسماء الملفات الصوتية
    - output_name: اسم الملف النهائي
    """
    combined = AudioSegment.empty()

    for audio_file in audio_files:
        if os.path.exists(audio_file):
            sound = AudioSegment.from_mp3(audio_file)
            combined += sound
            print(f"✅ تم إضافة: {audio_file}")
        else:
            print(f"❌ الملف غير موجود: {audio_file}")

    # حفظ الملف المدمج
    combined.export(output_name, format="mp3")
    print(f"✅ تم حفظ الملف المدمج: {output_name}")
    return output_name

print("✅ تم تعريف دالة دمج الملفات الصوتية")
```

```
def create_silence(duration_ms=1000):
    """إنشاء مقطع صامت بالملي ثانية"""
    return AudioSegment.silent(duration=duration_ms)

def merge_audio_with_silence(audio_files, output_name="podcast_final_pro.mp3", pause_ms=1500):
    combined = AudioSegment.empty()
    silence = create_silence(pause_ms)

    for i, audio_file in enumerate(audio_files):
        if os.path.exists(audio_file):
            sound = AudioSegment.from_mp3(audio_file)
            combined += sound
            if i < len(audio_files) - 1:
                combined += silence
            print(f"✅ تم إضافة: {audio_file}")

    combined.export(output_name, format="mp3")
    print(f"✨ تم الحفظ بنجاح: {output_name}")
    return output_name
```

### 🔊 ميزة جديدة: توحيد مستوى الصوت (Normalization)
هذه الدالة تضمن أن تكون جميع المقاطع الصوتية بنفس مستوى العلو لتجنب الاختلافات المزعجة بين صوت وآخر.

```
from pydub import effects

def normalize_audio(audio_segment, target_dbfs=-20.0):
    """
    تعديل مستوى الصوت ليكون موحداً بناءً على قيمة الديسيبل المستهدفة
    """
    # التصحيح: استخدام dBFS بدلاً من dbfs
    change_in_dbfs = target_dbfs - audio_segment.dBFS
    return audio_segment.apply_gain(change_in_dbfs)

print("✅ تم تصحيح دالة normalize_audio")
```

الآن، سنقوم بتعديل دالة الدمج لتستخدم ميزة توحيد الصوت تلقائياً.

```
def merge_audio_pro_normalized(audio_files, output_name="podcast_normalized.mp3", pause_ms=1500):
    combined = AudioSegment.empty()
    silence = AudioSegment.silent(duration=pause_ms)

    for i, audio_file in enumerate(audio_files):
        if os.path.exists(audio_file):
            sound = AudioSegment.from_file(audio_file)
            # توحيد الصوت قبل الإضافة
            normalized_sound = normalize_audio(sound)
            combined += normalized_sound

            if i < len(audio_files) - 1:
                combined += silence
            print(f"✅ تم توحيد ودمج: {audio_file}")

    combined.export(output_name, format="mp3")
    print(f"✨ تم حفظ الملف النهائي الموحد: {output_name}")
    return output_name
```

```
def add_background_music(podcast_path, music_path, output_name="final_podcast_with_music.mp3", music_volume_db=-25):
    """
    دمج موسيقى خلفية مع ملف البودكاست
    """
    if not os.path.exists(podcast_path):
        print(f"❌ ملف البودكاست غير موجود: {podcast_path}")
        return None

    podcast = AudioSegment.from_file(podcast_path)

    if os.path.exists(music_path):
        music = AudioSegment.from_file(music_path)
        # خفض صوت الموسيقى
        background = music + music_volume_db
        # دمج الموسيقى مع التكرار (loop) لتغطية كامل مدة البودكاست
        final_audio = podcast.overlay(background, loop=True)
        print(f"✅ تم دمج الموسيقى من: {music_path}")
    else:
        print(f"⚠️ لم يتم العثور على ملف الموسيقى {music_path}، سيتم حفظ الصوت الأصلي فقط.")
        final_audio = podcast

    final_audio.export(output_name, format="mp3")
    print(f"✨ تم الحفظ النهائي: {output_name}")
    return output_name
```

```
import numpy as np
from scipy.io import wavfile
from pydub import AudioSegment
import os

def generate_laugh_effect(filename="reaction_laugh.mp3"):
    # توليد صوت بسيط يشبه التنبيه أو الضحكة الإلكترونية
    sample_rate = 44100
    duration = 0.5
    t = np.linspace(0, duration, int(sample_rate * duration))
    # مزيج من الترددات
    audio_data = (np.sin(2 * np.pi * 440 * t) + np.sin(2 * np.pi * 880 * t)) * 0.5
    audio_data = (audio_data * 32767).astype(np.int16)

    wav_name = "temp_laugh.wav"
    wavfile.write(wav_name, sample_rate, audio_data)

    # تحويل إلى MP3
    laugh = AudioSegment.from_wav(wav_name)
    laugh.export(filename, format="mp3")
    os.remove(wav_name)
    print(f"✅ تم إنشاء ملف مؤثر صوتي بديل باسم: {filename}")

generate_laugh_effect()
```

### 🎭 دمج المؤثرات المرفوعة في البودكاست
الآن سنقوم بتشغيل النسخة النهائية التي تستخدم الملفات التي قمت برفعها بدلاً من الصمت.

```
!pip install -q gTTS pydub
import os
from gtts import gTTS
from pydub import AudioSegment
from IPython.display import Audio, display
from google.colab import files

def text_to_speech_ar(text, filename):
    tts = gTTS(text=text, lang='ar', slow=False)
    tts.save(filename)
    return filename

def merge_audio_with_silence(audio_files, output_name="sudanese_podcast_final.mp3", pause_ms=1300):
    combined = AudioSegment.empty()
    silence = AudioSegment.silent(duration=pause_ms)
    for i, audio_file in enumerate(audio_files):
        if os.path.exists(audio_file):
            sound = AudioSegment.from_mp3(audio_file)
            combined += sound
            if i < len(audio_files) - 1:
                combined += silence
    combined.export(output_name, format="mp3")
    return output_name

def run_sudanese_pro_podcast_v2():
    custom_scripts = [
        ("intro", "يا أخوانا سلام عليكم، مرحب بيكم في قعدتنا السودانية الليلة.", 1500),
        ("reaction_laugh", "REACTION", 500),
        ("topic", "الليلة دايرين نتكلم عن الشاي باللبن والونسة السمحة في بيوتنا.", 1500),
        ("outro", "شكراً ليكم كَتير، ونشوفكم المرة الجاية في أمان الله.", 1000)
    ]
    audio_files = []
    print("🔄 جاري معالجة الصوت والنبرات السودانية...")
    for file_id, text, pause in custom_scripts:
        filename = f"{file_id}.mp3"
        if text == "REACTION" and os.path.exists(filename):
            audio_files.append(filename)
            print(f"✅ تم دمج المؤثر: {filename}")
        elif text != "REACTION":
            text_to_speech_ar(text, filename)
            audio_files.append(filename)

    final_path = merge_audio_with_silence(audio_files)
    print(f"✨ تم الإنتاج بنجاح: {final_path}")
    display(Audio(final_path))
    files.download(final_path)

run_sudanese_pro_podcast_v2()
```

## 🚀 ميزة متقدمة: إنتاج بودكاست كامل من ملف سكريبت
هذه الميزة تتيح لك قراءة ملف `script.txt` يحتوي على الحوار، وتحويله إلى بودكاست مع إضافة موسيقى خلفية.

```
import os
from pydub import AudioSegment

def create_advanced_podcast(script_file='script.txt', bg_music='rababa.mp3'):
    final_audio = AudioSegment.empty()

    # إنشاء ملف تجريبي إذا لم يكن موجوداً
    if not os.path.exists(script_file):
        with open(script_file, 'w', encoding='utf-8') as f:
            f.write("host1|يا هلا بيكم في حلقتنا الليلة.\nhost2|ومرحب بكل المتابعين من كل مكان.")

    print(f"📖 جاري معالجة السكريبت من: {script_file}")

    with open(script_file, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if '|' in line:
                voice_id, text = line.strip().split('|', 1)
                temp_file = f"temp_{i}.mp3"
                text_to_speech_ar(text, temp_file)
                if os.path.exists(temp_file):
                    voice_part = AudioSegment.from_file(temp_file)
                    final_audio += voice_part + AudioSegment.silent(duration=1000)

    # إضافة موسيقى خلفية إذا وجدت
    if os.path.exists(bg_music):
        print(f"🎸 دمج موسيقى الخلفية: {bg_music}")
        bg = AudioSegment.from_file(bg_music)
        # تقليل صوت الخلفية ليكون خافتاً
        final_audio = final_audio.overlay(bg - 25, loop=True)

    output_path = "Final_Script_Podcast.mp3"
    final_audio.export(output_path, format="mp3")
    print(f"✅ تم الإنتاج النهائي: {output_path}")
    return output_path

# لتشغيل الميزة الجديدة:
# result = create_advanced_podcast()
# display(Audio(result))
```

```
# تجربة الميزة المتقدمة: إنتاج بودكاست من سكريبت
from IPython.display import Audio, display

# تشغيل الدالة وإنتاج الملف
result_path = create_advanced_podcast()

# عرض النتيجة للاستماع
if os.path.exists(result_path):
    display(Audio(result_path))
else:
    print("❌ فشل إنشاء ملف البودكاست المتقدم.")
```

### 🎻 الإنتاج المتقدم مع موسيقى الربابة
هذا الكود يقوم بربط الدرايف واستخدام الملفات الموجودة في مجلد المشروع لدمج الحوار مع الموسيقى.

```
from google.colab import drive, files
import os
from pydub import AudioSegment
from IPython.display import Audio, display

# إعداد المسارات مع خاصية التحقق من وجود الدرايف
drive_base = '/content/drive/MyDrive/Podcast_Project'
local_base = '/content/Podcast_Project'

# تحديد المسار النشط بناءً على توفر الدرايف
if os.path.exists('/content/drive/MyDrive'):
    base_path = drive_base
    print("✅ سيتم الحفظ في Google Drive")
else:
    base_path = local_base
    print("⚠️ سيتم الحفظ محلياً (Fallback to Local) لتعذر ربط الدرايف")

sfx_path = os.path.join(base_path, 'SFX')
script_path = os.path.join(base_path, 'script.txt')

def create_podcast_with_rababa():
    if not os.path.exists(base_path): os.makedirs(base_path)
    if not os.path.exists(sfx_path): os.makedirs(sfx_path)

    final_audio = AudioSegment.empty()

    if not os.path.exists(script_path):
        print(f"📝 جاري إنشاء ملف سكريبت تجريبي في: {script_path}")
        with open(script_path, 'w', encoding='utf-8') as f:
            f.write("host|يا هلا بيكم في بودكاست الربابة السودانية.\nguest|ومرحب بكل الناس السامعانا.")

    with open(script_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if '|' in line:
                voice_id, text = line.strip().split('|', 1)
                temp_name = f"voice_{i}.mp3"
                text_to_speech_ar(text, temp_name)
                if os.path.exists(temp_name):
                    part = AudioSegment.from_file(temp_name)
                    final_audio += part + AudioSegment.silent(duration=1000)

    # دمج الربابة (من المجلد أو من الملف المحلي إذا وجد)
    sfx_file = os.path.join(sfx_path, 'rababa.mp3')
    if not os.path.exists(sfx_file) and os.path.exists('rababa.mp3'):
        sfx_file = 'rababa.mp3'

    if os.path.exists(sfx_file):
        print(f"🎸 تم العثور على الربابة ({sfx_file})، جاري الدمج...")
        sfx = AudioSegment.from_file(sfx_file)
        final_audio = final_audio.overlay(sfx - 20, loop=True)
    else:
        print("ℹ️ لم يتم العثور على موسيقى خلفية، سيتم الإنتاج بدونها.")

    output_file = os.path.join(base_path, 'Final_Podcast_Pro.mp3')
    final_audio.export(output_file, format="mp3")
    print(f"✅ تم إنشاء البودكاست: {output_file}")
    display(Audio(output_file))

    if base_path == local_base:
        files.download(output_file)

# محاولة التشغيل
try:
    if not os.path.exists('/content/drive/MyDrive'):
        drive.mount('/content/drive')
    create_podcast_with_rababa()
except Exception as e:
    print(f"📡 ملاحظة: استمرار العمل بدون درايف بسبب: {e}")
    create_podcast_with_rababa()
```
